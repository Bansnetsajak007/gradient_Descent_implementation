```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
housing_data =  pd.read_csv('house-price-prediction-using-linear-regression.txt', usecols = [0,2],names=['size','price'])

housing_data.head()
```

```python
X = housing_data['size'].values/1000
Y = housing_data['price'].values/1000
```

```python
X = (X - np.mean(X)) / np.std(X)
Y = (Y - np.mean(Y)) / np.std(Y)
```

```python
n = len(Y)
n
```

```python
# initialize parameters
b0 = 0     # Intercept
w1 = 0     # Weight for area_of_house (x1)
learning_rate = 0.01  # Learning rate
epochs = 1000   # Number of iterations
```

```python
for epoch in range(epochs):
    y_pred = b0 + w1 * X
    errors = Y - y_pred

    # Compute gradients
    db0 = -2 * np.sum(errors)
    dw1 = -2 * np.sum(errors * X)

    # Update parameters
    b0 -= learning_rate * db0
    w1 -= learning_rate * dw1

    if epoch % 100 == 0:
        rss = np.mean(errors ** 2)
        print(f"Epoch {epoch}: RSS = {rss:.4f} | b0 = {b0:.4f}, w1 = {w1:.4f}")


```

```python
# Plotting the results
plt.scatter(X, Y, color='blue', label='Data Points')
plt.scatter(X, b0 + w1 * X, color='red', label='Predicted Points')    
plt.plot(X, b0 + w1 * X, color='green', label='Regression Line')    
plt.xlabel('Size (Standardized)')
plt.ylabel('Price (Standardized)')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.show()
```
